```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
from IPython.display import display as ds
```

```python
#LOAD .CSV FILE
df = pd.read_csv("C:/Users/vikas/OneDrive/Desktop/aus_ccx_21.csv")
df
```

```python
#SEE / EXPLORE THE DATA
df.dtypes
```

```python
df.shape
```

```python
#REMOVAL THE LEADING OR TRAILING SPACE FROM ALL STRING COLUMNS
df[df.select_dtypes(include = "object").columns] = df[df.select_dtypes(include = "object").columns].astype("string").map(str.strip, na_action = "ignore")
```

```python
df.dtypes
```

```python
# 3. EMAIL CHANGE TO LOWER CASE, VALIDATE, REPLACE INVALID WITH NaN
df["email"] = df["email"].str.strip().str.lower()
valid_email = df["email"].str.match(r"^[\w\.-]+@[\w\.-]+\.\w+$", na = False)
df.loc[~ valid_email, "email"] = np.nan
df

```

```python
df.columns
```

```python
# 4. Dates Correction and Replace to NaN ( as per email) 
df["join_date"] = df["join_date"].replace("Not Available", np.nan)
df["join_date"] = pd.to_datetime(df["join_date"],
                                 errors = "coerce",
                                 dayfirst = True)
df
```

```python
# 5 . Salary remove $ and USD change to empty string , change to numeric
df["salary"] = df["salary"].replace(r"[\$USD\s]", "", regex = True)
df["salary"] = pd.to_numeric(df["salary"],
                             errors = "coerce")
print(ds(df))
print(df.dtypes)
```

```python
# 5 . Remarks Col -  remove trailing and leading space , also double and single space between the string
df["remarks"] = df["remarks"].mask(df["remarks"].isna(), np.nan)
df["remarks"] = df["remarks"].str.strip()
df["remarks"] = df["remarks"].str.replace("  ","", regex = False)
df
```

```python
# 6. Sales = remove $ and leave <NA> as it is.
df["sales"] = df["sales"].replace(r"[\$,]","", regex = True)
df["sales"] = pd.to_numeric(df["sales"],
                            errors = "coerce")
df
```

```python
df.dtypes
```

```python
#7. Time Column - remove 'hrs' and change to numeric. 
df["time_productive"] = df["time_productive"].replace(r"hrs","", regex = True)
df["time_productive"] = pd.to_numeric(df["time_productive"],
                                     errors = "coerce")
df
```

```python
# 8 . Remove Duplicates (first_name , last_name , email) 
df = df.drop_duplicates(subset = ["email"]).reset_index(drop = True)
df = df.drop_duplicates(subset = ["first_name", "last_name"])
df
```

```python
df.shape
```

```python
# 9 . Placeholder remarks - unknown , email - unknown@no_mail.com
df["remarks"] = df["remarks"].fillna("unknown")
df["email"] = df["email"].fillna("unknown@no_mail.com")
```

```python
df
```

```python
df = df.reset_index(drop = True)
df
```

```python
# 10. SAVE TO aus_ccx_cln.csv 
df.to_csv("aus_ccx_21_cln.csv", index = False)
```

## BUSINESS INSIGHTS / REQUIREMENT

```python
print("The highest recorded sales is:",df["sales"].max())
print("The central average is:", df["sales"].mean())
print("The middle point of distribution is:", df["sales"].median())
print("The most frequently observed value is:", df["sales"].mode()[0])
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
max_sales = df["sales"].max()
max_row = df[df["sales"] == max_sales]
plt.figure(figsize=(12,6))
sns.barplot(x=df.index, y=df["sales"], color="skyblue")
sns.barplot(x=max_row.index, y=max_row["sales"], color="red")
for idx, value in zip(max_row.index, max_row["sales"]):
    plt.text(idx, value + 50, f"{value}", ha='center', color='red')
plt.title("Sales with Highest Sale Highlighted")
plt.xlabel("Index")
plt.ylabel("Sales")
plt.show()
```

```python
mean_sales = df["sales"].mean()
print("The central average is:", mean_sales)
plt.figure(figsize=(12,6))
sns.barplot(x=df.index, y=df["sales"], color="lightgreen")
plt.axhline(mean_sales, color='purple', linestyle='--', linewidth=2, label=f"Mean: {mean_sales:.2f}")
plt.title("Sales with Central Average Highlighted")
plt.xlabel("Row Index")
plt.ylabel("Sales")
plt.legend()
plt.show()
```
